```python
!pip install pandas networkx matplotlib

import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt

# Load CSV file into a DataFrame
df = pd.read_csv("/content/Founds_Network short.csv")

# Initialize an empty graph
G = nx.Graph() # you can use nx.DiGraph() instead of nx.Graph() to create a directed graph.

# Add edges to the graph from the source and target columns
G.add_edges_from(zip(df['Source'], df['Target']))

# Draw the graph to visualize it
nx.draw(G, with_labels=True)
plt.show()
```

#Spring Layout
Simulates a “spring” force between nodes, often providing a natural and visually pleasing spread.

```python
pos = nx.spring_layout(G, k=3, seed=42)  # Adjust `k` for more/less spacing
nx.draw(G, pos=pos, with_labels=True)
```

#Circular Layout
Places nodes in a circular arrangement, which can help reduce clustering in smaller networks.

```python
pos = nx.circular_layout(G)
nx.draw(G, pos=pos, with_labels=True)
```

# Shell Layout
Arranges nodes in concentric circles, which can help with visualization of hierarchical or layered structures.

```python
pos = nx.shell_layout(G)
nx.draw(G, pos=pos, with_labels=True)
```

# Random Layout

Positions nodes randomly, which may reduce clustering but is less controlled and can sometimes be messy.

```python
pos = nx.random_layout(G)
nx.draw(G, pos=pos, with_labels=True)
```

#Increase Node Size and Edge Width (for Larger Graphs)

For large graphs, it might also help to adjust the node size and edge width to make individual nodes and edges more visible:

```python
nx.draw(G, pos=pos, with_labels=True, node_size=500, width=0.5)
```
